```python

data = datasets.load_breast_cancer()
X, y = data.data, data.target

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1234)

# Further split the training data into training and cross-validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1234)
```

```python
def accuracy(y_test, y_pred):
    return np.sum(y_test == y_pred) / len(y_test)
```

```python
tot_feature = X.shape[1]
max_tree = 30

max_acc = 0
best_model = None

opt_n_tree = None
opt_n_feature  = None

for n_tree in range(2, max_tree+1):
  for n_feature in range(2, tot_feature+1):
    clf = RandomForest(n_trees=n_tree, n_features=n_feature)
    clf.fit(X_train, y_train)

    pred_val = clf.predict(X_val)
    acc = accuracy(y_val, pred_val)

    if acc > max_acc:
      max_acc = acc
      best_model = clf
      opt_n_tree = n_tree
      opt_n_feature = n_feature

print(f"CV accuracy for best mode: {max_acc}")
print(f"Optimal number of trees: {opt_n_tree}")
print(f"Optimal number of features: {opt_n_feature}")

pred_test = clf.predict(X_test)
acc = accuracy(y_test, pred_test)
print(f"test accuracy for best mode: {acc}")

```
